# **Detect and Treat Outliers**

```python
# Identifying outliers using the IQR method
Q1 = data_cleaned1.quantile(0.25)
Q3 = data_cleaned1.quantile(0.75)
IQR = Q3 - Q1

# Filter outliers
data = data_cleaned1[~((data_cleaned1 < (Q1 - 1.5 * IQR)) | (data_cleaned1 > (Q3 + 1.5 * IQR))).any(axis=1)]
```
